# Manipulating array elements with basic operations

```python
# These operations apply to each element in the arrays.

array_one = np.array([10,20,30,40])
array_two = np.array([1, 2, 3,4])

# Subtraction
array_three = array_one - array_two
print(array_three)

# Multiplication
array_four = array_one * array_two
print(array_four)
```

# Indexing Arrays

```python
# One dimensional arrays are similar to lists
array_one = np.array([4,7,1,4])
array_one[3]

# array_one at index three is the value 4
```

# Integer array indexing

```python
# Two dimensiona arrays are indexed this way.
array_one = np.array([[1,2], [3, 4]])
print(array_one) # notice the shape and positions of the numbers

# the output is like this [[1 2]
#############             [3 4]]
# to print the 3 in this array, do this:
print(array_one[1,0])  # Remember, the 1 is the row, and the 0 is the column.  So, 1,0 returns the 
                       # value in the spot of row 1, column 0 - which is 3.  Also, remember, Python indexing starts at 0.
```

# Slicing

```python
# Slicing allows us to select parts of an array
array_one = np.array([3,4,5,7,9])
print(array_one[:3])  # here we 'slice' out the values from index 0 to index 2.  This can be confusing.
                      #  Just remember that Python indexing goes UP TO the last number, but doesn't include it.  
                      #  For example, index [:3] returns the values in index spots 0,1,2 which are 3,4,5
```

```python
# or slice the 2nd and 3rd values with 1:3.  Confused yet?  
print(array_one[1:3])  # This returns the values 4,5. The 4 is in index 1, and the 5 is in index 2.
```

# Multi dimiensional array indexing

```python
#multidimensional indexing
array_one = np.array([[3,4,5,6], [7,8,9,10]])
array_one
```

```python
# to slice the 1st and 2nd values from the 3rd column which are 5 and 9
print(array_one[0:2,2])  # select 0:2 rows and 2 column.  this slices 5 and 9 out of rows 0,1 and column 2
```
